```python
def lighten_color(color, amount=0.4):
    """
    Lightens the given color by multiplying (1-luminosity) by the given amount.
    Input can be matplotlib color string, hex string, or RGB tuple.

    Parameters:
    - color (str or tuple): Input color.
    - amount (float): Amount to lighten by. A value of 0 will give the original color, and 1 will give white.

    Returns:
    - tuple: Lightened color in RGB format.
    """
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2])
```

```python
# Assuming a second set of Top@10 scores for a different setting for each model
top_50_scores_setting_1 = [0.899018484023499, 0.643466112623585, 0.7598868032669437, 0.8252973205330276]
top_50_scores_setting_2 = [0.8828825597245139, 0.8498457565105101, 0.8216514814549107, 0.8851424061984361] # Example alternative settings
# Adding a third setting to the plot for each model
# top_10_scores_setting_3 = [0.7580957157185844, 0.655896260209199, 0.6626307493910303, 0.6969479868175956]  # Example scores for third setting

colors_setting_1 = ['blue', 'green', 'red', 'purple']
width=0.35
models = ['Multi-VAE', 'MostPop', 'ItemKNN', 'PureSVD']
x = np.arange(len(models))  # the label locations

colors_setting_2_corrected = [lighten_color(c) for c in colors_setting_1]
# colors_setting_3_corrected = [lighten_color(c, amount=0.2) for c in colors_setting_1]  # Lighten colors further for the third setting
import matplotlib.pyplot as plt
import numpy as np

# Correcting the missing import for Line2D
from matplotlib.lines import Line2D

fig, ax = plt.subplots(figsize=(14, 8))

# Calculate the new width for three bars
width = width / 1.5

# Plot bars for each setting
for i, model in enumerate(models):
    ax.bar(x[i] - width, top_50_scores_setting_1[i], width, color=colors_setting_1[i], label=f'{model} - Setting 1')
    ax.bar(x[i], top_50_scores_setting_2[i], width, color=colors_setting_2_corrected[i], label=f'{model} - Setting 2')
    # ax.bar(x[i] + width, top_10_scores_setting_3[i], width, color=colors_setting_3_corrected[i], label=f'{model} - Setting 3')

# Adjust labels and titles
ax.set_xlabel('Models')
ax.set_ylabel('Scores at Top@50')
ax.set_title('Hit Result of Top@50 for Different Models')
ax.set_xticks(x)
ax.set_xticklabels(models)

# Update the legend to reflect three settings
legend_elements = [
    Line2D([0], [0], color='gray', lw=4, label='Without bounding box'),
    Line2D([0], [0], color=lighten_color('gray') , lw=4, label='With bounding box'),
    # Line2D([0], [0], color='gray', lw=4, linestyle=':', label='Setting 3')
]
legend_elements.extend([Line2D([0], [0], color=colors_setting_1[i], lw=4, label=models[i]) for i in range(len(models))])

ax.legend(handles=legend_elements, bbox_to_anchor=(1.05, 1), loc='upper left')

ax.grid(axis='y')
# plt.ylim(0.3, 0.8)

fig.tight_layout()

plt.show()
```
